## Challenge 01

```python
def filtro(x):
    if x < 1000:
        y = 1
    elif x >= 1000 and x < 10000:
        y = 2
    elif x >= 10000 and x < 100000:
        y = 3
    elif x >= 100000 and x < 1000000:
        y = 4
    else:
        y = 5
    return y
```

```python
imdb['bin'] = imdb.apply(lambda imdb: filtro(imdb['Votes']), axis=1)
imdb.head()
```

## Challenge 02

```python
def media(x,y):
    return x/y
```

```python
imdb['Rev/min'] = imdb.apply(lambda imdb: media(imdb['Revenue (Millions)'], imdb['Runtime (Minutes)']), axis=1)
imdb.head()
```

## Challenge 03

```python
def nuevoRanking(x,y):
    if 'Comedy' in y:
        x -= 1
    if 'Thriller' in y:
        x += 1
    return x
```

```python
imdb['New Ranking'] = imdb.apply(lambda imdb: nuevoRanking(imdb['Rating'], imdb['Genre']), axis=1)
imdb.head()
```

## Challenge 04

```python
def challenge4(x,y):
    z = 0
    
    for i in x:
        z += ord(i)
    z /= y
    z = int(z)
    
    primo = 'Es primo'
    
    if z != 0:
        for n in range(2,z):
            if z % n == 0:
                primo = 'No es primo'
    else:
        primo = 'No es primo'
    
    
                    
    return primo
```

```python
imdb['Es primo'] = imdb.apply(lambda imdb: challenge4(imdb['Title'], imdb['Votes']), axis=1)
imdb.head(25)
```
